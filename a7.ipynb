```python
import statistics

grades = []
a = 0
while (True):
    grade = int(input("Enter a test score: "))
    if grade < 0:
        break

    if grade >= 90:
        a += 1
    grades.append(grade)
print("Number of A's:", a)
print("Average:", statistics.mean(grades)) # or sum(grades) / len(grades)
```

```python
pairs = []
for i in range(100000, 999999 + 1 - 20):
    s_i = str(i)
    if (s_i == s_i[::-1]):
        for j in range(i + 1, i + 20): # less than 20 apart
            s_j = str(j)
            if (s_j == s_j[::-1]):
                pairs.append([i, j])

print(pairs)
```

```python
nums = []
for i in range(1, 10001):
    num = i
    digits = []
    while i > 0:
        digits.append(i % 10)
        i = i // 10
    
    mul = 1
    for digit in digits: mul *= digit

    if (num == sum(digits) + mul):
        nums.append(num)

print(nums)
```
